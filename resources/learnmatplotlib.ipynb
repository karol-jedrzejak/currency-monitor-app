```python
day = [1,2,3,4,5,6,7,8]
orgp = [20,30,40,20,30,50,60,70]
newp = [21,33,37,18,32,56,67,67]
```

```python
plt.plot(day,orgp,'b-o',label='original price');
plt.plot(day,newp,'r-o',label='new price');
plt.legend()
plt.title("Simple line plot");
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.savefig("pred.png",dpi=300)
```

```python
plt.subplot(3,1,1)
plt.plot(day,orgp,'b-o',label='original price');
plt.legend()
plt.title("Simple line plot");
plt.xlabel("Oś X")
plt.ylabel("Oś Y")

plt.subplot(3,1,2)
plt.plot(day,newp,'r-o',label='new price');
plt.legend()
plt.title("Simple line plot");
plt.xlabel("Oś X")
plt.ylabel("Oś Y")

diff = [o - p for o, p in zip(orgp,newp)]
plt.subplot(3,1,3)
plt.bar(day,diff,color="green",label='diff');
plt.legend()
plt.title("Simple line plot");
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
```
